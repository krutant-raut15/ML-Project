```python
#### Create a dataset with missing value
dict_ = {
    "Name":['Krutant','Akhil','Tushar','Savita'],
    "Gender":['Male','Male',None,'Female'],
    "Age":[22,np.nan,24,np.nan]

}

data_miss = pd.DataFrame(dict_)
data_miss
```

```python
### Handling Missing Value
data_miss.isnull().sum()
```

```python
data_miss['Gender'] = data_miss['Gender'].fillna(data_miss['Gender'].mode()[0])
```

```python
data_miss['Age']=data_miss['Age'].fillna(data_miss['Age'].mean())
```
